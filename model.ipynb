```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
train_data.head()
```

```python
train_data["learner_days_since_registration"] = scaler.fit_transform(train_data["learner_days_since_registration"].values.reshape(-1,1))
train_data["learner_cnt_other_courses_active"] = scaler.fit_transform(train_data["learner_cnt_other_courses_active"].values.reshape(-1,1))

train_data["learner_cnt_other_courses_items_completed"] = scaler.fit_transform(train_data["learner_cnt_other_courses_items_completed"].values.reshape(-1,1))
train_data["learner_cnt_other_courses_paid_items_completed"] = scaler.fit_transform(train_data["learner_cnt_other_courses_paid_items_completed"].values.reshape(-1,1))
train_data["learner_other_revenue"] = scaler.fit_transform(train_data["learner_other_revenue"].values.reshape(-1,1))

train_data["subscription_period_order"] = scaler.fit_transform(train_data["subscription_period_order"].values.reshape(-1,1))
train_data["days_since_last_payment"] = scaler.fit_transform(train_data["days_since_last_payment"].values.reshape(-1,1))

train_data["days_til_next_payment_due"] = scaler.fit_transform(train_data["days_til_next_payment_due"].values.reshape(-1,1))
train_data["cnt_days_active_before_payment_period"] = scaler.fit_transform(train_data["cnt_days_active_before_payment_period"].values.reshape(-1,1))

train_data["cnt_days_since_last_activity"] = scaler.fit_transform(train_data["cnt_days_since_last_activity"].values.reshape(-1,1))
```

```python
y_train = train_data['is_retained']
train_data = train_data.drop(['is_retained'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(train_data,y_train,test_size=0.3,shuffle=True)
```

```python
logreg0 = LogisticRegression()
grid= {"C": np.logspace(-3,3,7), "penalty":["l1","l2"]}
logreg_cv = GridSearchCV(logreg0,grid,cv=10)
logreg_cv.fit(xtrain, ytrain)
print("Best parameters of Logistic Regression :",logreg_cv.best_params_)
```

```python
logreg = LogisticRegression(C=logreg_cv.best_params_["C"] , penalty=logreg_cv.best_params_["penalty"])
logreg = logreg.fit(xtrain, ytrain) 
#print("Logistic Regression Accuracy :",logreg.score(test_x,test_y))
logreg_prediction = logreg.predict(xtest)
accuracy_score(ytest, logreg_prediction)
```
